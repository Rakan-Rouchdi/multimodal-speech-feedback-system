```python
import os
import numpy as np
import librosa
import matplotlib.pyplot as plt
import seaborn as sns
import random
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Bidirectional, LSTM, Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.regularizers import l2

# Reproducibility
SEED = 42
np.random.seed(SEED)
random.seed(SEED)
tf.random.set_seed(SEED)

# Fix for deprecated NumPy attributes used by librosa
if not hasattr(np, 'complex'):
    np.complex = complex

# Dataset path
DATASET_PATH = "Audio"

# Emotion mapping
emotion_map = {
    "01": "neutral", "02": "calm", "03": "happy", "04": "sad",
    "05": "angry", "06": "fearful", "07": "disgust", "08": "surprised"
}

# Augmentation functions
def noise(data):
    noise_amp = 0.035 * np.random.uniform() * np.amax(data)
    return data + noise_amp * np.random.normal(size=data.shape[0])

def stretch(data, rate=0.85):
    try:
        stft = librosa.stft(data)
        stft_stretched = librosa.phase_vocoder(D=stft, rate=rate, hop_length=512)
        return librosa.istft(stft_stretched, hop_length=512)
    except:
        return data

def shift(data):
    shift_range = int(np.random.uniform(low=-5, high=5) * 1000)
    return np.roll(data, shift_range)

def pitch(data, sr, pitch_factor=0.7):
    return librosa.effects.pitch_shift(y=data, sr=sr, n_steps=pitch_factor)

# Feature extraction (MFCC + Delta + ZCR + Chroma + RMSE)
def extract_features(y, sr, max_len=250):
    try:
        y_trimmed, _ = librosa.effects.trim(y, top_db=40)

        mfcc = librosa.feature.mfcc(y=y_trimmed, sr=sr, n_mfcc=13)
        delta = librosa.feature.delta(mfcc)
        delta2 = librosa.feature.delta(mfcc, order=2)
        zcr = librosa.feature.zero_crossing_rate(y_trimmed)
        chroma = librosa.feature.chroma_stft(y=y_trimmed, sr=sr)
        rmse = librosa.feature.rms(y=y_trimmed)

        features = np.vstack([mfcc, delta, delta2, zcr, chroma, rmse]).T  # (time, features)

        if features.shape[0] < max_len:
            pad_width = max_len - features.shape[0]
            features = np.pad(features, ((0, pad_width), (0, 0)), mode='constant')
        else:
            features = features[:max_len, :]

        return features
    except Exception as e:
        print(f"❌ Feature extraction error: {e}")
        return None

# Load original files
def extract_originals(dataset_path):
    X, y, paths = [], [], []
    for root, _, files in os.walk(dataset_path):
        for file in files:
            if file.endswith(".wav"):
                parts = file.split('.')[0].split('-')
                emotion = emotion_map.get(parts[2])
                path = os.path.join(root, file)
                try:
                    y_raw, sr = librosa.load(path, sr=None)
                    features = extract_features(y_raw, sr)
                    if features is not None:
                        X.append(features)
                        y.append(emotion)
                        paths.append(path)
                except:
                    continue
    return np.array(X), y, paths

# Augment only training set
def augment_and_extract_training(paths, emotions):
    AUGMENTATIONS = ['noise', 'stretch', 'shift', 'pitch']
    X_aug, y_aug = [], []
    for path, label in zip(paths, emotions):
        try:
            y_raw, sr = librosa.load(path, sr=None)

            # original
            features = extract_features(y_raw, sr)
            if features is not None:
                X_aug.append(features)
                y_aug.append(label)

            for aug in AUGMENTATIONS:
                if aug == 'noise':
                    y_augmented = noise(y_raw)
                elif aug == 'stretch':
                    y_augmented = stretch(y_raw)
                elif aug == 'shift':
                    y_augmented = shift(y_raw)
                elif aug == 'pitch':
                    y_augmented = pitch(y_raw, sr)

                features = extract_features(y_augmented, sr)
                if features is not None:
                    X_aug.append(features)
                    y_aug.append(label)
        except:
            continue
    return np.array(X_aug), y_aug

# Load and split
X_orig, y_orig, paths = extract_originals(DATASET_PATH)
le = LabelEncoder()
y_encoded = le.fit_transform(y_orig)
y_cat = to_categorical(y_encoded)

# Split into Train (80%) + Temp (20%)
X_train_val, X_test, y_train_val, y_test = train_test_split(X_orig, y_cat, test_size=0.1, stratify=y_encoded, random_state=SEED)
X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val, test_size=0.1111, stratify=np.argmax(y_train_val, axis=1), random_state=SEED)

# Augment only training set
train_idx = [i for i in range(len(X_orig)) if any((X_orig[i] == x).all() for x in X_train)]
train_paths = [paths[i] for i in train_idx]
y_train_labels = [y_orig[i] for i in train_idx]

X_train_aug, y_train_aug = augment_and_extract_training(train_paths, y_train_labels)
y_train_encoded = le.transform(y_train_aug)
y_train_cat = to_categorical(y_train_encoded)

print("✅ Train:", X_train_aug.shape, y_train_cat.shape)
print("✅ Val  :", X_val.shape, y_val.shape)
print("✅ Test :", X_test.shape, y_test.shape)

# Deep CNN + BiLSTM
input_shape = X_train_aug.shape[1:]

model = Sequential([
    Conv1D(64, 5, activation='relu', input_shape=input_shape, kernel_regularizer=l2(0.001)),
    BatchNormalization(),
    MaxPooling1D(2),
    Dropout(0.3),

    Conv1D(128, 3, activation='relu', kernel_regularizer=l2(0.001)),
    BatchNormalization(),
    MaxPooling1D(2),
    Dropout(0.3),

    Bidirectional(LSTM(64, return_sequences=False, kernel_regularizer=l2(0.001))),
    Dropout(0.4),

    Dense(128, activation='relu', kernel_regularizer=l2(0.001)),
    Dropout(0.3),

    Dense(y_train_cat.shape[1], activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Callbacks
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True, verbose=1)
checkpoint = ModelCheckpoint("best_model.h5", monitor='val_accuracy', save_best_only=True, verbose=1)

# Train
history = model.fit(
    X_train_aug, y_train_cat,
    validation_data=(X_val, y_val),
    epochs=50,
    batch_size=32,
    callbacks=[early_stop, checkpoint]
)

# Accuracy Plot
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Val')
plt.title("Model Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.show()

# Evaluation (on test set)
y_pred_proba = model.predict(X_test)
y_pred = np.argmax(y_pred_proba, axis=1)
y_true = np.argmax(y_test, axis=1)

y_pred_labels = le.inverse_transform(y_pred)
y_true_labels = le.inverse_transform(y_true)

print("📊 Classification Report:")
print(classification_report(y_true_labels, y_pred_labels, digits=3))

# Confusion Matrix
cm = confusion_matrix(y_true_labels, y_pred_labels, labels=le.classes_)
plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=le.classes_, yticklabels=le.classes_)
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.tight_layout()
plt.show()

# Inference on new file
def predict_emotion_from_file(path):
    y_raw, sr = librosa.load(path, sr=None)
    features = extract_features(y_raw, sr)
    if features is None:
        return "❌ Error loading features"
    features = np.expand_dims(features, axis=0)
    prediction = model.predict(features)
    predicted_label = le.inverse_transform([np.argmax(prediction)])
    return predicted_label[0]
```
